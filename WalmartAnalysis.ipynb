```python
#Dependencies

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load - these are saved csv's
file_to_load = "Walmart.csv"
file_to_load1 = "kaggle_income.csv"
```

```python
#Walmart dataframe
walmart_pd = pd.read_csv(file_to_load, encoding="utf-8")
walmart_pd.head()
```

```python
#Income dataframe
income_pd = pd.read_csv(file_to_load1, encoding="utf-8")
income_pd.head()
```

```python
#Split the Store Name into address fields

walmart_pd['Store Name'] = walmart_pd['Store Name'].str.replace(' - Walmart.com','')
walmart_pd[['Store Type', 'Address', 'City', 'State Zip']] = walmart_pd['Store Name'].str.split(",", n = 3, expand = True)
walmart_pd[['B','State','Zip_Code']] = walmart_pd['State Zip'].str.split(" ",n=2, expand=True)
walmart_pd = walmart_pd.drop(columns=['Store Name','State Zip','B'])

walmart_pd.head()
```

```python
#Get rid of any trailing or leading zeros for merging later

walmart_pd[['Zip_Code']] = walmart_pd['Zip_Code'].str.strip()
walmart_pd.head()
```

```python
#keep the last 5 digits- makes sure that the zip code is 5 numbers
walmart_pd['Zip_Code'] = walmart_pd['Zip_Code'].str[-5:]
```

```python
walmart_pd.head()
```

```python
#Character to numeric
walmart_pd['Zip_Code']=walmart_pd['Zip_Code'].astype(int)
```

```python
walmart_pd.head()
```

```python
#Merge by zip code
walinc_table_pd = pd.merge(walmart_pd, income_pd, on="Zip_Code", how="inner")

#print first few observations of new dataframe
walinc_table_pd.head()
```

```python

#Keep only the variables we need 
walmart_new_pd = walinc_table_pd[['Store Number', 'Store Type', 'Address', 'City_x', 'State', 'Zip_Code', 'Median']]
walmart_new_pd.head()
```
